```python
from tensorflow.keras.models import load_model
import numpy as np
import json
```

```python
model = load_model('Model1.h5')
data = np.load('features_labels_testset_noAugment.npz')
features = data['features']
labels = data['labels']

with open('label_dict.json', 'r') as f:
    label_dict = json.load(f)

print(np.array(features[0]).shape)
```

```python
single_feature = np.array(features[1])
single_feature_reshaped = single_feature.reshape(1, single_feature.shape[0], 1)

prediction = model.predict(single_feature_reshaped)

predicted_class = np.argmax(prediction)
print(predicted_class)

reverse_label_dict = {v: k for k, v in label_dict.items()}
actual_label = reverse_label_dict[predicted_class]
print(actual_label)
```

```python
# Initialize a counter for correct predictions
correct_predictions = 0

# Convert labels to integer format for comparison with predicted class
int_labels_test = [label_dict[label] for label in labels]

for i, feature in enumerate(features):
    # Reshape the feature to (1, 40, 1)
    feature = feature.reshape(1, feature.shape[0], 1)

    # Predict the class
    prediction = model.predict(feature)
    predicted_class = np.argmax(prediction)

    # Check if the prediction is correct and if so, increment the counter
    if predicted_class == int_labels_test[i]:
        correct_predictions += 1

# Calculate the accuracy
accuracy = correct_predictions / len(features)

print(f"Accuracy: {accuracy * 100:.2f}%")
```
