```python
def display_mfccs(file_name):
    """
    This function takes in the path for an audio file as a string, loads the audio,
    and extracts MFCC features from the audio. Then it displays the MFCCs.
    """
   
    try:
        # Load the audio file using librosa.load(). This returns the audio data and the sampling rate.
        # We're using 'kaiser_fast' as the resampling method for performance reasons.
        audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast') 
        
        # Extract MFCCs from the audio data using librosa.feature.mfcc().
        # y: the audio time series
        # sr: the sampling rate
        # n_mfcc: the number of MFCCs to compute (here, we're computing 40 MFCCs)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
     
    except Exception as e:
        print("Error encountered while parsing file: ", file_name, "Exception: ", str(e))
    return None 
 
     
    # Create a new figure for plotting
    plt.figure(figsize=(10, 4))

    # Use librosa.display.specshow() to display the MFCCs as a heatmap.
    # This function is similar to plt.imshow(), but it does some additional formatting 
    # to better display spectrograms.
    # The x-axis will represent time, and the y-axis will represent the different MFCCs.
    # The color of each point will represent the value of a particular MFCC at a particular time.
    librosa.display.specshow(mfccs, x_axis='time')
    
    # Display a colorbar next to the plot
    plt.colorbar()
    
    # Set the title of the plot
    plt.title('MFCC')
    
    # Make sure the layout is tight (i.e., no wasted space)
    plt.tight_layout()
    
    # Display the plot
    plt.show()
```

```python
def extract_features(file_name):
    """
    This function takes in the path for an audio file as a string, loads the audio,
    and extracts MFCC features from the audio.
    """
   
    try:
        audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast') 
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        mfccs_processed = np.mean(mfccs.T,axis=0)
     
    except Exception as e:
        print("Error encountered while parsing file: ", file_name, "Exception: ", str(e))
    return None 

    
    display_mfccs(file_name)
    
    return mfccs_processed
```

```python
# Set the path to the full dataset 
fulldatasetpath = r'C:\Users\786me\Desktop\Kashmiri Translation\downloads\WAV'

metadata = pd.DataFrame(columns=['filename','label'])

# Let's read the audio files and extract MFCCs
features = []
```

```python
for subdir, dirs, files in os.walk(fulldatasetpath):
    for file in files:
        file_path = os.path.join(subdir, file)
        class_label = os.path.splitext(file)[0]  # assuming label is the filename without the extension
        data = extract_features(file_path)
        features.append([data, class_label])
        metadata = metadata.append({'filename': file, 'label': class_label}, ignore_index=True)

# Convert into a Panda dataframe 
featuresdf = pd.DataFrame(features, columns=['feature','class_label'])

print('Finished feature extraction from ', len(featuresdf), ' files')
```
