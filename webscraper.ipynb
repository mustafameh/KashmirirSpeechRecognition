```python
import os
import requests
from bs4 import BeautifulSoup
```

```python

# URL of the page
url = "https://dsal.uchicago.edu/cgi-bin/app/hassan_query.py?qs=a&searchhws=yes"

# Fetch the page content
response = requests.get(url)
response.raise_for_status()  # Ensure we got a valid response

# Parse the content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all div elements with class 'hw_result'
result_divs = soup.find_all('div', class_='hw_result')

# Download directory
download_dir = "downloads"

# Ensure download directory exists
os.makedirs(download_dir, exist_ok=True)
```

```python

# Iterate over the result divs
for div in result_divs:
    # Get the audio URL from the source tag inside the div
    audio_url = div.find('source')['src']

    # Get the title text from the div text
    title_text = div.text.strip().split('\n')[0]

    # Use title as filename, replace spaces with underscores and append .mp3
    filename = f"{title_text.replace(' ', '_')}.mp3"

    # Download the audio file and save it to the download directory
    audio_response = requests.get(audio_url)
    audio_response.raise_for_status()  # Ensure we got a valid response

    with open(os.path.join(download_dir, filename), 'wb') as f:
        f.write(audio_response.content)

    print(f"Downloaded {filename} from {audio_url}")

```

```python
# Base URL of the page without the query string
base_url = "https://dsal.uchicago.edu/cgi-bin/app/hassan_query.py?qs={}&searchhws=yes"

# Download directory
download_dir = "downloads"

# Ensure download directory exists
os.makedirs(download_dir, exist_ok=True)
```

```python
from string import ascii_lowercase
# Loop over all lowercase letters a-z
for letter in ascii_lowercase:
    # Create the URL for the current letter
    url = base_url.format(letter)

    # Fetch the page content
    response = requests.get(url)
    response.raise_for_status()  # Ensure we got a valid response

    # Parse the content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all div elements with class 'hw_result'
    result_divs = soup.find_all('div', class_='hw_result')

    # Iterate over the result divs
    for div in result_divs:
        # Get the audio URL from the source tag inside the div
        audio_url = div.find('source')['src']

        # Get the title text from the div text
        title_text = div.text.strip().split('\n')[0]

        # Use title as filename, replace spaces with underscores and append .mp3
        filename = f"{title_text.replace(' ', '_')}.mp3"

        # Download the audio file and save it to the download directory
        audio_response = requests.get(audio_url)
        audio_response.raise_for_status()  # Ensure we got a valid response

        with open(os.path.join(download_dir, filename), 'wb') as f:
            f.write(audio_response.content)

        print(f"Downloaded {filename} from {audio_url}")
```
